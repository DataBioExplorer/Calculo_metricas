```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from google.colab import files

#  Upload da imagem
uploaded = files.upload()
image_path = list(uploaded.keys())[0]  # Obtm o nome do arquivo enviado
img = Image.open(image_path)

#  Converter para escala de cinza
gray = img.convert("L")

#  Converter para preto e branco (binarizao)
threshold = 128  # Define o nvel de corte
bw = gray.point(lambda x: 255 if x > threshold else 0, mode="1")

#  Exibir as imagens lado a lado
plt.figure(figsize=(15, 5))

# Imagem original
plt.subplot(1, 3, 1)
plt.imshow(img)
plt.title("Imagem Original")
plt.axis("off")

# Escala de cinza
plt.subplot(1, 3, 2)
plt.imshow(gray, cmap="gray")
plt.title("Escala de Cinza")
plt.axis("off")

# Preto e branco
plt.subplot(1, 3, 3)
plt.imshow(bw, cmap="gray")
plt.title("Preto e Branco (Binarizada)")
plt.axis("off")

plt.show()
```
